```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('cal_housing_clean.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
x_data = df[['housingMedianAge', 'totalRooms', 'totalBedrooms', 'population',
       'households', 'medianIncome']]
y_values = df[['medianHouseValue']]
```

```python
x_data.head()
```

```python
y_values.head()
```

```python
X_train, X_eval, y_train, y_eval = train_test_split(x_data, y_values, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = pd.DataFrame(data = scaler.transform(X_train), columns = X_train.columns, index = X_train.index)
```

```python
X_train.head()
```

```python
X_test = pd.DataFrame(data = scaler.transform(X_train), columns = X_train.columns, index = X_train.index)
```

```python
df.columns
```

```python
husing_age = tf.feature_column.numeric_column('housingMedianAge')
tot_rooms = tf.feature_column.numeric_column('totalRooms')
tot_bedrms = tf.feature_column.numeric_column('totalBedrooms')
pop = tf.feature_column.numeric_column('population')
households = tf.feature_column.numeric_column('households')
mednincom = tf.feature_column.numeric_column('medianIncome')
```

```python
feat_colms = [husing_age, tot_rooms, tot_bedrms, pop, households, mednincom]
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x = X_train, y = y_train, batch_size = 10, num_epochs = 1000, shuffle = True)
```

```python
dnn_model = tf.estimator.DNNRegressor(hidden_units = [6, 6, 6], feature_columns = feat_colms)
```

```python
dnn_model.train(input_fn = input_func, steps = 1000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(x=X_eval, y=y_eval, batch_size=10, num_epochs=1000, shuffle=False)
```

```python
dnn_model.evaluate(input_fn = eval_input_func, steps = 1000)
```

```python
pred_input_func = tf.estimator.inputs.pandas_input_fn(x=X_eval, batch_size=10, num_epochs=1, shuffle=False)
```

```python
dnn_model.predict(input_fn=pred_input_func)
```

```python
predictions = list(dnn_model.predict(input_fn = pred_input_func))
```

```python
predictions
```

```python
final_pred = []
for pred in predictions:
    final_pred.append(pred['predictions'])
```

```python
final_pred
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_eval, final_pred)**0.5
```
